```python
import numpy as np
import pandas as pd

import bar_chart_race as bcr
```

```python
# Data import
df = pd.read_csv("data/data.csv", index_col='date')
```

```python
df.head()
```

```python
# Calculating active cases of each province
df['province1_active'] = df['province1_pos'] - df['province1_dis']
df['province2_active'] = df['province2_pos'] - df['province5_dis']
df['bagmati_province_active'] = df['bagmati_province_pos'] - df['bagmati_province_dis']
df['gandaki_province_active'] = df['gandaki_province_pos'] - df['gandaki_province_dis']
df['province5_active'] = df['province5_pos'] - df['province5_dis']
df['karnali_province_active'] = df['karnali_province_pos'] - df['karnali_province_dis']
df['sudurpaschim_province_active'] = df['sudurpaschim_province_pos'] - df['sudurpaschim_province_dis']
```

```python
# Selecting on the active cases column
df_active = df[['province1_active','province2_active','bagmati_province_active','gandaki_province_active','province5_active','karnali_province_active','sudurpaschim_province_active']]
df_active.head()
```

```python
# Plotting the graph

bcr.bar_chart_race(
    df=df_active,
    filename='visualization/covid_19_cases.mp4',
    orientation='h',
    sort='desc',
    n_bars=7,
    fixed_order=False,
    fixed_max=True,
    steps_per_period=10,
    interpolate_period=False,
    label_bars=True,
    bar_size=.95,
    period_label={'x': .99, 'y': .25, 'ha': 'right', 'va': 'center'},
    period_summary_func=lambda v, r: {'x': .99, 'y': .18,
                                      's': f'Total Active Cases: {v.nlargest(7).sum():,.0f}',
                                      'ha': 'right', 'size': 8, 'family': 'Courier New'},
    perpendicular_bar_func=None,
    period_length=500,
    figsize=(5, 3),
    dpi=144,
    cmap='dark12',
    title='COVID-19 Active Cases by Province',
    bar_label_size=7,
    tick_label_size=7,
    shared_fontdict={'family' : 'Helvetica', 'color' : '.1'},
    scale='linear',
    writer=None,
    fig=None,
    bar_kwargs={'alpha': .7},
    filter_column_colors=False)
```
